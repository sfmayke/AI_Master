```python
from typing import Dict, Iterable, List
import torch

class CharTokenizer:
    """
    A deterministic character-level tokenizer.

    Guarantees:
    - Token IDs are contiguous integers starting from 0
    - Vocabulary order is stable
    - Unknown characters map to <UNK>
    - Encoding and decoding are inverse operations (except for <UNK>)
    """
    
    PAD_TOKEN = "<PAD>"
    UNK_TOKEN = "<UNK>"
    
    def __init__(self, vocabulary: List[str]) -> None:
        if len(vocabulary) != len(set(vocabulary)):
            raise ValueError("Vocabulary contains duplicate tokens.")
            
        self._token_id_for_char: Dict[str, int] = {char: idx for idx, char in enumerate(vocabulary)}
        self._char_for_token_id: Dict[int, str] = {idx: char for idx, char in enumerate(vocabulary)}
        self.pad_token_id = self._token_id_for_char[self.PAD_TOKEN]
        self.unk_token_id = self._token_id_for_char[self.UNK_TOKEN]
            
    @classmethod
    def train_from_text(cls, text: str) -> "CharTokenizer":
        """
        Build a tokenizer from raw text.

        Vocabulary is:
        [<PAD>, <UNK>, sorted(unique characters)]
        """
        unique_chars = sorted(set(text))
        vocabulary = [cls.PAD_TOKEN, cls.UNK_TOKEN, *unique_chars]
        return cls(vocabulary)

    def encode(self, text: str) -> torch.Tensor:
        """
        Convert text into a tensor of token IDs.
        Unknown characters are mapped to <UNK>.
        """
        token_ids = [self._token_id_for_char.get(char, self.unk_token_id) for char in text]
        return torch.tensor(token_ids, dtype=torch.long)

    def decode(self, token_ids: Iterable[int]) -> str:
        """
        Convert token IDs back into text.
        Special tokens are preserved as literal strings.
        """
        chars = []
        for token_id in map(int, token_ids):
            if token_id not in self._char_for_token_id:
                raise ValueError(f"Invalid token id: {token_id}")

            chars.append(self._char_for_token_id[token_id])
        return "".join(chars)

    @property
    def vocabulary_size(self) -> int:
        return len(self._token_id_for_char)
```

```python
tokenizer = CharTokenizer.train_from_text(text)
```

```python
print(tokenizer.encode("Hello world"))
print(tokenizer.decode(tokenizer.encode("Hello world")))
```

```python
print(f"Vocabulary size: {tokenizer.vocabulary_size}")
```

```python
from torch.utils.data import Dataset
import torch


class TokenIdsDataset(Dataset):
    """
    Dataset that produces (input, target) token ID sequences
    for next-token prediction.

    Given a 1D tensor of token IDs:
        data = [t0, t1, t2, ..., tN]

    Each item is:
        x = data[pos : pos + block_size]
        y = data[pos + 1 : pos + block_size + 1]

    Guarantees:
    - All returned sequences have length == block_size
    - Targets are inputs shifted by one position
    """

    def __init__(self, data: torch.Tensor, block_size: int) -> None:
        if not isinstance(data, torch.Tensor):
            raise TypeError("data must be a torch.Tensor")

        if data.ndim != 1:
            raise ValueError("data must be a 1D tensor of token IDs")

        if block_size <= 0:
            raise ValueError("block_size must be a positive integer")

        if len(data) <= block_size:
            raise ValueError(
                "data length must be greater than block_size"
            )

        self.data = data
        self.block_size = block_size

    def __len__(self) -> int:
        """
        Number of valid starting positions.

        Last valid start is:
            len(data) - block_size - 1
        """
        return len(self.data) - self.block_size

    def __getitem__(self, pos: int):
        if not isinstance(pos, int):
            raise TypeError("Index must be an integer")

        if pos < 0 or pos >= len(self):
            raise IndexError("Index out of range")

        x = self.data[pos : pos + self.block_size]
        y = self.data[pos + 1 : pos + self.block_size + 1]

        return x, y
```

```python
# Step 2 - Tokenize the Text
encoded_text = tokenizer.encode(text)
dataset = TokenIdsDataset(encoded_text, 64)
```

```python
from torch.utils.data import DataLoader, RandomSampler

# RandomSampler allows to read random items from a datasset
sampler = RandomSampler(dataset, replacement=True)
# Dataloader will laod two random samplers using the sampler
dataloader = DataLoader(dataset, batch_size=2, sampler=sampler)
```

**Building the Multi-Head Attention Block in PyTorch**

In this video, we will build the multi-head attention block, a crucial component of the Transformer model, from scratch using PyTorch.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from typing import Dict


class AttentionHead(nn.Module):
    """
    Single causal self-attention head.

    Shape notation:
        B = batch size
        T = sequence length (number of tokens)
        E = embedding dimension
        H = attention head dimension

    Input:
        x: (B, T, E)

    Output:
        y: (B, T, H)
    """

    def __init__(self, config: Dict) -> None:
        super().__init__()

        embed_dim: int = config["embedding_dim"]   # E
        head_dim: int = config["head_size"]        # H
        dropout: float = config["dropout_rate"]
        use_bias: bool = config.get("use_bias", True)

        if embed_dim <= 0 or head_dim <= 0:
            raise ValueError("embedding_dim and head_size must be positive")

        # Per-token projections; no cross-token interaction
        self.q_proj = nn.Linear(embed_dim, head_dim, bias=use_bias)
        self.k_proj = nn.Linear(embed_dim, head_dim, bias=use_bias)
        self.v_proj = nn.Linear(embed_dim, head_dim, bias=use_bias)

        self.dropout = dropout

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """
        Args:
            x: Input embeddings of shape (B, T, E)

        Returns:
            Contextualized embeddings of shape (B, T, H)
        """
        # Project embeddings to Q, K, V
        q = self.q_proj(x)  # Query: what each token looks for
        k = self.k_proj(x)  # Key: what each token offers
        v = self.v_proj(x)  # Value: information carried forward

        # Causal attention with optional dropout (training only)
        return F.scaled_dot_product_attention(
            q,
            k,
            v,
            is_causal=True,
            dropout_p=self.dropout if self.training else 0.0,
        )
```

```python
class MultiHeadSelfAttention(nn.Module):
    def __init__(self, config: Dict) -> None:
        super().__init__()

        embed_dim: int = config["embedding_dim"]
        head_dim: int = config["head_size"]
        num_heads: int = config["num_heads"]
        dropout: float = config["dropout_rate"]
        use_bias: bool = config.get("use_bias", True)

        if embed_dim != num_heads * head_dim:
            raise ValueError("embedding_dim must equal num_heads * head_size")

        head_config = {
            "embedding_dim": embed_dim,
            "head_size": head_dim,
            "dropout_rate": dropout,
            "use_bias": use_bias,
        }

        # Independent attention heads
        self.heads = nn.ModuleList(
            [AttentionHead(head_config) for _ in range(num_heads)]
        )

        # Mix information across heads
        self.out_proj = nn.Linear(embed_dim, embed_dim, bias=use_bias)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # Apply each head and concatenate along feature dimension
        concat = torch.cat([head(x) for head in self.heads], dim=-1)
        return self.out_proj(concat)
```

```python
# ---- config ----
config = {
    "embedding_dim": 64,
    "head_size": 32,
    "dropout_rate": 0.1,
    "use_bias": True,
}

# ---- device ----
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ---- instantiate ----
attn = AttentionHead(config).to(device)
attn.eval()  # inference mode (no dropout)

# ---- random input ---- 
batch_size = 2
seq_len = 10
embedding_dim = config["embedding_dim"]

x = torch.randn(batch_size, seq_len, embedding_dim, device=device)

# ---- forward pass ----
with torch.no_grad():
    y = attn(x)

# ---- checks ----
print("Input shape :", x.shape)
print("Output shape:", y.shape)
print("Device      :", y.device)
```
